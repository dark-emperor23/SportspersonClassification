```python
import os
from icrawler.builtin import GoogleImageCrawler

# Main dataset folder
main_folder = "Sports_Dataset"

# Names of sportspeople (folder names + search keywords)
people = {
    "Cristiano Ronaldo": "Cristiano Ronaldo",
    "Virat Kohli": "Virat Kohli Indian cricketer",
    "Rafal Nadal": "Rafael Nadal tennis player",
    "Maria Sharapova": "Maria Sharapova tennis player",
    "Neeraj Chopra": "Neeraj Chopra javelin throw",
    "Rohit Sharma": "Rohit Sharma Indian cricketer"
}

def get_image_count(folder):
    return len([f for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

# Ensure at least 500 images per person
MIN_IMAGES = 500
MAX_ATTEMPTS = 3  # to avoid infinite crawling

for person, keyword in people.items():
    folder_path = os.path.join(main_folder, person)
    os.makedirs(folder_path, exist_ok=True)

    print(f"\nüîç Downloading images for: {person}")

    attempt = 0
    while get_image_count(folder_path) < MIN_IMAGES and attempt < MAX_ATTEMPTS:
        crawler = GoogleImageCrawler(storage={"root_dir": folder_path})
        crawler.crawl(keyword=keyword, max_num=700)  # Request more than 500
        attempt += 1
        current_count = get_image_count(folder_path)
        print(f"  Attempt {attempt}: Downloaded {current_count} images so far...")

    if get_image_count(folder_path) >= MIN_IMAGES:
        print(f"‚úÖ {person}: Reached at least 500 images.")
    else:
        print(f"‚ö†Ô∏è {person}: Only {get_image_count(folder_path)} images after {MAX_ATTEMPTS} attempts.")

print("\nüèÅ All images downloaded (500+ per person where possible).")

```
